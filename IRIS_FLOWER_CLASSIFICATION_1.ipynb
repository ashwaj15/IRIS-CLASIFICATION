# 3. **Data preparation**


---
*  To Define 'x' and 'y'. 


> 'x' contains all the input variables such as independent features, and 'y' should contain the dependent variable which is dependent on independent variables, the output.







```python
x=df.iloc[:,:4]
y=df.iloc[:,4]
```

To View 'x'

```python
x
```

To View 'y'

```python
y
```



---



# 4. **Training the algorithm**






---




> Import train_test_split to split the data into train and test datasets.






```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```


* To View their shapes. Use the shape method to view.

#5. **Creating the Model (Classification)**



---



*  So here we are going to classify the Iris flowers dataset using 
logistic regression. For creating the model, import LogisticRegression from the sci-kit learn library.

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```



*   Now train the model using the fit method. In the fit method, pass training datasets in it. x_train and y_train are the training datasets.


```python
model.fit(x_train,y_train)
```



---



#6.  **Making Predictions**


---







> Now we predict the results using predict method.



```python
y_pred=model.predict(x_test)
```

```python
y_pred
```



---



*  To Find the accuracy of the model and view the confusion matrix.


---












```python
from sklearn.metrics import accuracy_score,confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
accuracy=accuracy_score(y_test,y_pred)*100
print("Accuracy of the model is {:.2f}".format(accuracy))
```
